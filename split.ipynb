```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Cargar los datos desde el archivo CSV con la codificación adecuada
file_path = '/mnt/ssd/media/liglesias/codigo_beca_colab/DICOM_patientIDs.csv'
df = pd.read_csv(file_path, delimiter=';', encoding='latin1')

# Definir los rangos para las categorías de fracción de eyección
def categorize_ejection_fraction(ef_value):
    if pd.isna(ef_value):
        return 'unknown'
    elif ef_value < 30:
        return 'low'
    elif ef_value < 55:
        return 'medium'
    else:
        return 'high'

# Aplicar la categorización a las columnas de fracción de eyección
df['EjectionFraction_RV_Category'] = df['EjectionFraction RV'].apply(categorize_ejection_fraction)
df['EjectionFraction_LV_Category'] = df['EjectionFraction LV'].apply(categorize_ejection_fraction)

# Filtrar los datos donde 'Segmentation' sea 'Sí'
training_si = df[df['Segmentation'] == 'Sí']

# Verificar la cantidad de instancias por combinación de categorías
comb_counts = training_si.groupby(['PatientSex', 'EjectionFraction_RV_Category', 'EjectionFraction_LV_Category']).size()
print("Combinaciones de categorías en el conjunto completo:")
print(comb_counts)

# Filtrar combinaciones con al menos dos instancias
valid_combs = comb_counts[comb_counts >= 2].index
filtered_training_si = training_si[
    training_si.set_index(['PatientSex', 'EjectionFraction_RV_Category', 'EjectionFraction_LV_Category']).index.isin(valid_combs)
]

# Realizar el split asegurando balance de género y categorías de fracción de eyección
train, test = train_test_split(
    filtered_training_si, 
    test_size=13, 
    stratify=filtered_training_si[['PatientSex', 'EjectionFraction_RV_Category', 'EjectionFraction_LV_Category']], 
    random_state=42
)

# Comprobar que los grupos están balanceados
def check_balance(df, label):
    counts = df.groupby(['PatientSex', 'EjectionFraction_RV_Category', 'EjectionFraction_LV_Category']).size()
    print(f"\nCombinaciones de categorías en el conjunto {label}:")
    print(counts)
    return counts

train_counts = check_balance(train, "de entrenamiento")
test_counts = check_balance(test, "de prueba")

# Comprobar que las distribuciones son similares
print("\nDistribución en el conjunto de entrenamiento:")
print(train_counts / len(train))

print("\nDistribución en el conjunto de prueba:")
print(test_counts / len(test))
```

```python
import os
import numpy as np
root_dir = '/mnt/nfs/home/liglesias/data_tgv'

training_dir_tgv = os.path.join(root_dir, "train")
testing_dir_tgv = os.path.join(root_dir, "test")

assert os.path.exists(training_dir_tgv), f"El directorio {training_dir_tgv} no existe"
assert os.path.exists(testing_dir_tgv), f"El directorio {testing_dir_tgv} no existe"

def list_nifti_files(directory):
    nifti_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".nii") or file.endswith(".nii.gz"):
                nifti_files.append(os.path.join(root, file))
    return nifti_files

training_files_tgv = list_nifti_files(training_dir_tgv)
testing_files_tgv = list_nifti_files(testing_dir_tgv)

print(f"Se encontraron {len(training_files_tgv)} archivos en {training_dir_tgv}")
print(f"Se encontraron {len(testing_files_tgv)} archivos en {testing_dir_tgv}")
```

```python
all_files_tgv = training_files_tgv + testing_files_tgv
images_tgv = [nib.load(file).get_fdata() for file in all_files_tgv]
image_shapes_tgv = [img.shape for img in images_tgv]
print(image_shapes_tgv)
average_shape = np.mean(image_shapes_tgv, axis=0).astype(int)
print(f"Tamaño promedio: {average_shape}")
min_shape_tgv = np.min(image_shapes_tgv, axis=0)
max_shape_tgv = np.max(image_shapes_tgv, axis=0)

print(len(all_files_tgv))
print(f"Tamaño mínimo: {min_shape_tgv}")
print(f"Tamaño máximo: {max_shape_tgv}")
```

```python
# Extraer la profundidad (suponiendo que es la tercera dimensión)
depths_tgv = [shape[2] for shape in image_shapes_tgv]

# Calcular el número medio, mínimo y máximo de cortes
average_depth_tgv = int(np.mean(depths_tgv))  # Convertir a entero
min_depth_tgv = np.min(depths_tgv)
max_depth_tgv = np.max(depths_tgv)

# Imprimir resultados
print(f"Número total de estudios: {len(all_files_tgv)}")
print(f"Media del número de cortes por estudio: {average_depth_tgv}")
print(f"Número mínimo de cortes: {min_depth_tgv}")
print(f"Número máximo de cortes: {max_depth_tgv}")
```

```python
# Lista para almacenar los grosores de corte
slice_thicknesses = []

for file_path in all_files_tgv:
    # Cargar el archivo NIfTI
    nii_image = nib.load(file_path)
    header = nii_image.header

    # Extraer el grosor de corte (espaciado en la tercera dimensión)
    slice_thickness = header['pixdim'][3]  # pixdim[3] corresponde al eje Z
    slice_thicknesses.append(slice_thickness)

# Calcular estadísticas
average_thickness = np.mean(slice_thicknesses)
min_thickness = np.min(slice_thicknesses)
max_thickness = np.max(slice_thicknesses)

# Imprimir resultados
print(f"Número total de estudios: {len(all_files_tgv)}")
print(f"Media del grosor de corte (mm): {average_thickness:.2f}")
print(f"Grosor mínimo de corte (mm): {min_thickness:.2f}")
print(f"Grosor máximo de corte (mm): {max_thickness:.2f}")
```
