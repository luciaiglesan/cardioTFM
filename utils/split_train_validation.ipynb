```python
images_path = '/mnt/nfs/home/liglesias/nnUNet_raw/Dataset109_all_datasets/imagesTr'
labels_path = '/mnt/nfs/home/liglesias/nnUNet_raw/Dataset109_all_datasets/labelsTr'

all_image_files = [f for f in os.listdir(labels_path) if f.endswith(".nii.gz")]

sorted_list = sorted(all_image_files)

validation_ranges = list(range(1, 21)) + list(range(101, 149)) + list(range(341, 378))

val_patients = [f for f in sorted_list if int(f[:3]) in validation_ranges]
train_patients = [f for f in sorted_list if int(f[:3]) not in validation_ranges]

train_patients = [f.replace(".nii.gz", "") for f in train_patients]
val_patients = [f.replace(".nii.gz", "") for f in val_patients]

split_data = [
    {
        "train": train_patients,
        "val": val_patients
    }
]

output_file = "splits_final.json"

with open(output_file, "w") as json_file:
    json.dump(split_data, json_file, indent=4)

print(f"Archivo JSON guardado como {output_file}")
```
