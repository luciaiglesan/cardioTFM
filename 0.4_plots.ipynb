```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="white", rc=custom_params)

# Pallete: https://matplotlib.org/stable/_images/sphx_glr_named_colors_003.png
palette = {
    "Adelie": "rebeccapurple",
    "Chinstrap": "sienna",
    "Gentoo": "olivedrab"
}
```

```python
penguins = sns.load_dataset("penguins")

fig, axes = plt.subplots(1,2,figsize=(15, 5), width_ratios=[2, 1])

# Histogram
g = sns.histplot(data=penguins, x='flipper_length_mm', hue='species', 
             multiple='dodge', shrink=0.9, kde=True, palette=palette, ax=axes[0])
sns.move_legend(
    axes[0], "lower center",
    bbox_to_anchor=(0.75, 1), ncol=3, title=None, frameon=False,
)

# Violin plot
sns.violinplot(data=penguins, y='species', x='flipper_length_mm', hue='species', palette=palette, ax=axes[1])
axes[1].set_ylabel(None)
axes[1].set_yticklabels([])

plt.subplots_adjust(wspace=0.1)
plt.show()
```
